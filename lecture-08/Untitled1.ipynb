```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = Image.open("thor.jpg")
```

```python
arr = np.array(img)
```

```python
arr.shape
```

```python
X = arr.reshape(-1, 3)
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=8)
```

```python
model.fit(X)
```

```python
output = model.cluster_centers_[model.labels_]
```

```python
intimg = output.astype(np.uint8)
```

```python
convrted = intimg.reshape(arr.shape)
```

```python
plt.imshow(convrted)
```

```python
plt.imshow(arr)
```

```python
pallets = model.cluster_centers_.astype(int)
```

```python
plt.imshow(pallets.reshape([1, -1, 3]))
```
