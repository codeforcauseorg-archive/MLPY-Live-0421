```python
x = np.linspace(0, 2*np.pi, 100)
```

```python
sin = np.sin(x)
cos = np.cos(x)
tan = np.tan(x)
sq = x ** x
```

```python
plt.figure(figsize=(6, 6))

plt.subplot(2, 2, 4)
plt.plot(x, sin)

plt.subplot(2, 2, 2)
plt.plot(x, cos)

plt.subplot(2, 2, 3)
plt.plot(x, tan)

plt.subplot(2, 2, 1)
plt.plot(x, sq)
```

```python
plt.figure(figsize=(6, 6))

plt.subplot(2, 2, 1)
plt.plot(x, sin)

plt.subplot(2, 2, 2)
plt.plot(x, cos)

plt.subplot(2, 1, 2)
plt.plot(x, tan)

```

```python
img = np.zeros((1000, 1000), dtype=np.uint8)
```

```python
for r in range(img.shape[0]):
    for c in range(img.shape[1]):
        
        cx = r - (r%100)
        cy = c - (c%100)
        
        dist = np.sqrt(((r-cx)*(r-cx) + (c-cy)*(c-cy)))
        img[r, c] = dist
```

```python
img
```

```python
plt.imshow(img, cmap="gray")
```
