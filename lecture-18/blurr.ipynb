```python
from tensorflow.keras.preprocessing.image import load_img
import numpy as np
```

```python
img = load_img("../datasets/monestry.jpg")
```

```python
npimg = np.array(img)
```

```python
npimg.shape
```

```python
fil = np.ones([10, 10])
```

```python
res = np.empty((npimg.shape[0]-fil.shape[0] + 1, npimg.shape[1]-fil.shape[1] + 1, 3), dtype=np.uint8)
```

```python
for r in range(0, npimg.shape[0]-fil.shape[0] + 1):
    for c in range(0, npimg.shape[1]-fil.shape[1] + 1):
        
        chop = npimg[r:r+fil.shape[0], c:c+fil.shape[1]]
        res[r, c] = chop.mean(axis=0).mean(axis=0).astype(np.uint8)
```
